```python
import pandas as pd
import numpy as np
import os
```

```python
file_path = "oil/"
files = {}
for file in os.listdir(file_path):
    full_path = file_path+file
    if "csv" in full_path:
        files[f"{file}"] = pd.read_csv(full_path)
```

```python
files.keys()
```

```python
oil_df = pd.merge(files["OPEC production capacity.csv"],files["net_inventory_withdrawal.csv"], on = "Date", how="outer")
```

```python
oil_df = oil_df.merge(files["production.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["imports.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["OPEC production.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["inventory.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["exports.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["net_import.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["adjustment.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["Crude Oil and Liquid Fuels Supply.csv"],on="Date", how="outer")
oil_df = oil_df.merge(files["us imports from OPEC.csv"],on="Date", how="outer")
```

```python
oil_df.columns
```

```python
oil_df = oil_df[["Date","US Production","US Import","US Exports","US Net Import","Inventory",
                 "Adjustment","Import from OPEC","Crude Oil and Liquid Fuels Supply",
                 "OPEC Production Capacity","Net Inventory Withdrawals","OPEC Production"]]
oil_df = oil_df.dropna(how="any")
oil_df.head()
```

```python
oil_df = oil_df.merge(files["MacroTrends_Data_Download.csv"],on="Date", how="outer")
```

```python
oil_df = oil_df.dropna(how="any")
oil_df.head()
```

```python
oil_df.to_csv("us_crude_oil.csv")
```
